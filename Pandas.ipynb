```python
# doc du lieu qua nguon internet
irisdata = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/iris.csv',
           header = 0 ,sep = ',')
irisdata.head()
```

```python
# reshape data
# Lenh melt
irisdata.melt(['Name', 'PetalWidth'])
```

```python
# cac thong ke
# thong ke mo ta
irisdata.iloc[:, :4].describe()
```

```python
print('mean: ', irisdata['SepalLength'].mean())
print('max: ', irisdata['SepalLength'].max())
print('min: ', irisdata['SepalLength'].min())
print('standard deviation: ', irisdata['SepalLength'].std())
print('sum: ', irisdata['SepalLength'].sum())
print('median: ', irisdata['SepalLength'].median())
```

```python
print('sum: --------------------------', irisdata.iloc[:, :4].apply(sum))
# print('mean: --------------------------', irisdata.iloc[:, :4].apply(np.mean))
# print('max: --------------------------', irisdata.iloc[:, :4].apply(max))
# print('min: --------------------------', irisdata.iloc[:, :4].apply(min))
```

```python
def square(x):
    return x**2

irisdata.iloc[:,:4].apply(square).head()
```

```python
irisdata.iloc[:,:4].apply(lambda x: x**2).head()
```

```python
# thong ke theo nhom
irisdata.groupby(['Name']).sum()
# dimension, measurement
```

```python
# thong ke theo pivot_table

irisdata.pivot_table(columns = ['Name'], 
                     values = ['SepalLength', 'SepalWidth', 'PetalLength'], 
                     aggfunc = 'sum')
```

```python
# thong ke theo pivot_table

irisdata.pivot_table(
                     values = ['SepalLength', 'SepalWidth', 'PetalLength'], 
                     index = ['Name'], 
                     aggfunc = 'sum')
```

```python
# thong ke theo pivot_table

irisdata.pivot_table(values = ['SepalLength', 'SepalWidth', 'PetalLength'], 
                     index = ['Name'], 
                     aggfunc = {'SepalLength':sum,
                               'SepalWidth':max,
                               'PetalLength':np.mean})
```
